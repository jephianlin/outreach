#### 定理（代數基本定理）
任何用複數做為係數的 $n$ 次多項式方程  
恰有 $n$ 個複數解。

````python
f = x^2 + 1
factor(f)
````

````python
g = f.polynomial(QQ)
factor(g)
````

````python
g = f.polynomial(CC)
factor(g)
````

## 多項式
多項式可以寫成  
$ a_nx^n + a_{n-1}x^{n-1} + \cdots + a_0 $  
它的次方數都是非負的整數。

$\sqrt{x}$ 及 $\frac{1}{x}$ 都不是多項式。

多項式和整數一樣都可以**分解**（factor）

````python
p = x^2 - 4
factor(p)
````

````python
p = x^7 - 1
factor(p)
````

Sage 也可以計算多項式的**展開**（expand）

````python
p = (x+1)^10
expand(p)
````

### 符號表示式與多項式
在 Sage 中，`x` 是一個預設的未知數  
所以任何跟 `x` 有關的式子都用**符號表示式**（Symbolic Expression）來儲存  
實際上，許多數字（像是 $\pi$, $\sqrt{2}$ 等等）都是以符號來儲存

````python
type(x+1)
````

````python
type(pi)
````

````python
type(sqrt(2))
````

符號表示式只是方便用來儲存及計算  
而**多項式**則是一個比較完整的代數結構  
比如說，我們可以用 `(x+1).polynomial(QQ)`  
來得到 `x+1` 這個多項式  
這裡的 `QQ` 表示多項式的係數是有理數  
（大多時候我們遇到的多項式用有理數就足夠了）

````python
p = (x+1).polynomial(QQ)
type(p)
````

````python
p = x+1
p = p.polynomial(QQ)
type(p)
````

變成多項式以後就可以使用除法原理了：  
給定兩個多項式 $a(x)$ 及 $b(x)$  
存在唯一的**商式** $q(x)$ 及**餘式** $r(x)$  
使得 $a(x) = q(x)b(x) + r(x)$ 且 $\deg r(x) < \deg b(x)$

````python
a = (x+1)^5
a = expand(p)
a
````

````python
a = a.polynomial(QQ)
b = (x+1).polynomial(QQ)
c = (x^4 + x^3 + x^2 + x + 1).polynomial(QQ)
````

`a` 除以 `b` 餘式為 0

````python
a // b
````

````python
a % b
````

`a` 除以 `c` 等於  
`x+4` 餘 `5*x^3 + 5*x^2 - 3`

````python
a // c
````

````python
a % c
````

### 多項式與函數
多項式可以**代入**（substitute）數字

````python
p = x^2 + 5
p.subs(x=2)
````

這表示多項式有**函數**的性質

## 函數
函數就像機器一樣  
輸入一些東西以後  
會吐出一些東西

在 Sage 數定義一個函數  
（在程式設計數又叫做函式，以和數學中的函數做區別）
```Python
def function_name(var1, ..., var2=0, ...):
    do something
    return something
```

定義一個函數  
輸入一個數字 `x`  
吐出它的平方

````python
def f(x):
    return x^2
````

````python
f(3)
````

定義另一個函數  
輸入兩個數字 `x` 和 `p`  
吐出 `x` 的 `p` 次方

````python
def p_power(x,p=5):
    return x^p
````

````python
p_power(2,10)
````

定義函數時  
我們已經給定 `p` 的**預設值**為 `5`  
所以當 `p` 值沒輸入時  
函數會自動將 `p` 視為 `5`

````python
p_power(2)
````

實際上程式中的函數可以做更多事情：

判斷奇偶性（parity）

````python
def parity(x):
    if x % 2 == 0:
        print 'even'
    if x % 2 == 1:
        print 'odd'
````

````python
parity(3)
````

````python
parity(4)
````

計算 0 的個數

````python
def count_zeros(numbers):
    counter = 0
    for i in numbers:
        if i == 0:
            counter = counter + 1
    return counter
````

````python
a = [0,0,2,4,0,0,3]
count_zeros(a)
````
